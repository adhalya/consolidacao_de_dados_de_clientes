Um cliente entrou em contato comigo dizendo que precisa unificar 3 a 4 listas de clientes em um único arquivo Excel. Embora aproximadamente 70% dos registros coincidam em nome e NIF, há informações complementares que não podem ser perdidas. É essencial que o arquivo final não contenha duplicatas do mesmo cliente, garantindo que cada cliente apareça apenas uma vez, consolidando todas as informações disponíveis.

Para demonstrar como resolvi o seu problema, criei 3 pequenas planilhas que serão utilizadas no código a seguir.

```python
import pandas as pd
```

Importando arquivos .xlsx

```python
df_janeiro = pd.read_excel('/content/Sales_January.xlsx')
df_fevereiro = pd.read_excel('/content/Sales_February.xlsx')
df_marco = pd.read_excel('/content/Sales_March.xlsx')
```

Visualizando datasets

```python
df_janeiro.head()
```

```python
df_fevereiro
```

```python
df_marco
```

```python
df_concatenado = pd.concat([df_janeiro, df_fevereiro, df_marco], ignore_index=True)

df_final = (
    df_concatenado
    .sort_values("NIF")
    .groupby("NIF", as_index=False)
    .agg(lambda x: x.dropna().iloc[0] if not x.dropna().empty else None)
)
```

```python
df_final
```

Ordenando as colunas

```python
df_final = df_final[["NIF", "Cliente", "Email", "Telefone", "Morada", "Gestor_Conta", "Valor_Janeiro", "Valor_Fevereiro", "Valor_Marco"]]
```

```python
df_final.to_excel("Clientes.xlsx", index=False)
```
